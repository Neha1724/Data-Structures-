```python
import matplotlib.pyplot as plt      # Importing matplotlib for plotting
import numpy as np                   # Importing numpy for numerical operations
from scipy.integrate import solve_ivp  # Importing solve_ivp from scipy for solving initial value problems

plt.style.use('seaborn-poster')       # Setting the plot style

# Defining the differential equation dy/dt = cos(t)
F = lambda t, s: np.cos(t)

# Creating an array of time values for evaluation
t_eval = np.arange(0, 2*np.pi, 0.1)

# Solving the initial value problem using solve_ivp
# F is the function defining the differential equation
# [0, np.pi] is the time interval
# [0] is the initial condition
# t_eval is the array of time values for evaluation
sol = solve_ivp(F, [0, 2*np.pi], [0], t_eval=t_eval)

# Creating a figure for plotting
plt.figure(figsize=(12, 4))

# Plotting the solution of the differential equation
#plt.subplot(121)
plt.plot(sol.t, sol.y[0])
plt.xlabel('t')      # Labeling the x-axis
plt.ylabel('S(t)')   # Labeling the y-axis

# Plotting the difference between the solution and sin(t)
#plt.subplot(122)
#plt.plot(sol.t, sol.y[0] - np.sin(sol.t))
#plt.xlabel('t')                  # Labeling the x-axis
#plt.ylabel('S(t) - sin(t)')      # Labeling the y-axis

# Adjusting layout for better appearance
plt.tight_layout()

# Displaying the plot
plt.show()
```
